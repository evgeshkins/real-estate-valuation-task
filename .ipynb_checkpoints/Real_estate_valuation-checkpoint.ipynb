```python
def is_numeric_series(series):
    return pd.to_numeric(series, errors='coerce').notnull().all()

# Находим столбцы с типом данных object
object_cols = sold_objects.select_dtypes(include=['object']).columns

# Находим столбцы c типом данных object, которые можно преобразовать в числовой тип
num_cols_bool = sold_objects[object_cols].apply(is_numeric_series)
num_cols_bool
```

Как видим, значения, содержащие буквы, действительно присутствуют. Поработаем с каждым столбцом отдельно.

area_balcony

Как мы видим по результатам применения метода unique выше, в данном столбце очень много непонятных данных. Например: '3,5/2' - не совсем понятно, что значит данное значение: длина/ширина балкона или, может быть, арифметическая операция деления; или значение '7,8 3,8 2,2', вообще имеющее внутри себя три числа через пробел. Мы не фантазёры и поэтому оставим те значения, которые идеально парсятся к числовому типу данных, например: '3.0', а также в значениях по типу '10,6' заменим запятую на точку.

```python
# преобразуем значения, в которых есть запятая, так, чтобы вместо запятой была точка
import re
sold_objects['area_balcony'] = sold_objects['area_balcony'].apply(lambda x: str(x).replace(',', '.') if re.match(r'^\d*\.?\d*$', str(x).replace(',', '.')) else None)
sold_objects["area_balcony"].unique()
```

Данные стали немного чище. Но, как видим, присутствуют выбросы. Например, значение '9038.0'.

```python
# Преобразование значений столбца 'area_balcony' из строкового в числовой тип
sold_objects['area_balcony'] = sold_objects['area_balcony'].apply(pd.to_numeric, errors='coerce')
sold_objects["area_balcony"].unique()
```

Обработка выбросов

```python
area_balcony = sold_objects["area_balcony"]
# Определение среднего и стандартного отклонения
mean_val = area_balcony.mean()
std_dev = area_balcony.std()

# Определение границ для выбросов (например, убрать значения в пределах 3 стандартных отклонений от среднего)
lower_bound = mean_val - 3 * std_dev
upper_bound = mean_val + 3 * std_dev

# Замена выбросов на None
area_balcony_no_outliers = area_balcony.where((area_balcony >= lower_bound) & (area_balcony <= upper_bound), None)

print(area_balcony.describe())

print("Статистика после замены выбросов на None:")
print(area_balcony_no_outliers.describe())
```

```python
sold_objects["area_balcony"] = area_balcony_no_outliers
sold_objects.info()
```

Рассмотрим вещественные признаки, в которых могут быть выбросы

```python
float_cols = ['price', 'area_total', 'area_live', 'area_kitchen', 'area_balcony', 'keep_encoded']
```

```python
def replace_outliers_with_nan(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df.loc[~df[col].between(lower_bound, upper_bound), col] = np.nan

# Пройтись по всем столбцам датафрейма и заменить выбросы на NaN
for col in float_cols:
    replace_outliers_with_nan(X, col)
X.info()
```

Как видим, выбросов у нас было достаточно. Заполняем пропуски средними значениями.

```python
for col in float_cols:
    X[col].fillna(X[col].mean(), inplace=True)
```
